### Cleaning Data

```python
import pandas as pd
from google.colab import files

# 1. Unggah file CSV lagi
print("Silakan unggah file CSV:")
uploaded = files.upload()

# 2. Membaca file CSV yang baru diunggah
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# 3. Menampilkan beberapa data awal sebelum cleaning
print("\nðŸ”¹ Data Awal (5 Baris Pertama):")
print(df.head())

# 4. Mengecek jumlah missing values (data kosong)
print("\nðŸ”¹ Jumlah Missing Values Sebelum Cleaning:")
print(df.isnull().sum())

# 5. Mengisi missing values (pilihan: isi dengan 'Tidak Diketahui' atau median)
for col in df.columns:
    if df[col].dtype == "object":  # Jika kolom tipe string, isi dengan 'Tidak Diketahui'
        df[col].fillna("Tidak Diketahui", inplace=True)
    else:  # Jika kolom numerik, isi dengan median
        df[col].fillna(df[col].median(), inplace=True)

print("\nðŸ”¹ Jumlah Missing Values Setelah Cleaning:")
print(df.isnull().sum())

# 6. Menghapus data duplikat (jika ada)
df.drop_duplicates(inplace=True)
print("\nðŸ”¹ Jumlah Data Setelah Menghapus Duplikasi:", len(df))

# 7. Memeriksa dan mengonversi tipe data jika perlu
print("\nðŸ”¹ Informasi DataFrame Setelah Cleaning:")
print(df.info())

# Contoh: Jika ada kolom angka yang terbaca sebagai string, ubah ke numerik
numeric_columns = ["Total_Belanja"]  # Ubah sesuai dengan nama kolom yang ada
for col in numeric_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# 8. Membersihkan teks dari spasi berlebih atau karakter aneh
for col in df.select_dtypes(include=["object"]).columns:
    df[col] = df[col].str.strip()  # Hapus spasi di awal/akhir

# 9. Menampilkan data setelah cleaning
print("\nðŸ”¹ Data Setelah Cleaning (5 Baris Pertama):")
print(df.head())

# Simpan data yang telah dibersihkan
cleaned_filename = "cleaned_" + filename
df.to_csv(cleaned_filename, index=False)
print(f"\nâœ… Data telah dibersihkan dan disimpan sebagai '{cleaned_filename}'")
```

## Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# 1. Unggah file CSV
print("Silakan unggah file CSV:")
uploaded = files.upload()

# 2. Membaca file CSV yang diunggah
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# 3. Pastikan kolom yang relevan ada
print("\nKolom yang tersedia dalam dataset:")
print(df.columns)

# 4. Jumlah pelanggan unik
unique_customers = df["customer_unique_id"].nunique()
print(f"\nðŸ”¹ Jumlah pelanggan unik: {unique_customers}")

# 5. Visualisasi jumlah pelanggan per kota
plt.figure(figsize=(10, 5))
sns.countplot(y=df["customer_city"], order=df["customer_city"].value_counts().index[:10], palette="viridis")
plt.title("ðŸ”¹ Top 10 Kota dengan Jumlah Pelanggan Terbanyak")
plt.xlabel("Jumlah Pelanggan")
plt.ylabel("Kota")
plt.show()

# 6. Visualisasi jumlah pelanggan per state
plt.figure(figsize=(10, 5))
sns.countplot(y=df["customer_state"], order=df["customer_state"].value_counts().index, palette="coolwarm")
plt.title("ðŸ”¹ Jumlah Pelanggan per Negara Bagian")
plt.xlabel("Jumlah Pelanggan")
plt.ylabel("Negara Bagian")
plt.show()

# 7. Histogram jumlah transaksi per pelanggan
plt.figure(figsize=(8, 5))
df["customer_unique_id"].value_counts().hist(bins=30, color='skyblue', edgecolor='black')
plt.title("ðŸ”¹ Distribusi Jumlah Transaksi per Pelanggan")
plt.xlabel("Jumlah Transaksi")
plt.ylabel("Frekuensi")
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# @title Default title text
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# 1. Unggah file CSV
print("Silakan unggah file CSV:")
uploaded = files.upload()

# 2. Membaca file CSV yang diunggah
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# 3. Normalisasi nama kolom (menghapus spasi, mengubah ke huruf kecil)
df.columns = df.columns.str.strip().str.lower()

# 4. Menentukan kolom kota yang benar (berdasarkan daftar yang ditemukan)
possible_city_columns = ["customer_city", "kota", "city"]
found_city_column = None

for col in df.columns:
    if col in possible_city_columns:
        found_city_column = col
        break

if found_city_column:
    # 5. Menghitung jumlah pembeli per kota
    kota_terbanyak = df[found_city_column].value_counts().head(10)

    # 6. Menampilkan daftar kota dengan pembeli terbanyak
    print("\nðŸ”¹ 10 Kota dengan Jumlah Pembeli Terbanyak:")
    print(kota_terbanyak)

    # 7. Membuat chart kota dengan pembeli terbanyak
    plt.figure(figsize=(12, 6))
    sns.barplot(x=kota_terbanyak.index, y=kota_terbanyak.values, palette="viridis")
    plt.xlabel("Kota", fontsize=12)
    plt.ylabel("Jumlah Pembeli", fontsize=12)
    plt.title("10 Kota dengan Jumlah Pembeli Terbanyak", fontsize=14)
    plt.xticks(rotation=45)

    # 8. Menampilkan nilai di atas batang
    for i, v in enumerate(kota_terbanyak.values):
        plt.text(i, v + 5, str(v), ha='center', va='bottom', fontsize=10)

    # 9. Menampilkan chart
    plt.show()

else:
    print("\nâš ï¸ Kolom yang berisi data kota tidak ditemukan dalam dataset.")
    print("ðŸ” Kolom yang tersedia:", df.columns.tolist())
```
